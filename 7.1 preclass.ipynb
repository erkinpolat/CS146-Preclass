```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
binomial = stats.binom(200000, 0.5)
```

```python
samples = stats.binom(200000000, 0.5).rvs(size=1000)
```

```python
plt.hist(samples)
```

```python
normalapprox = stats.norm(200000000, 1000000)
```

```python
x = np.linspace(-20000, 20000)
plt.plot(x, normalapprox.pdf(x))
```
